# Create The Dataset Of Match Data from Match IDs

```python
#our dataframe design
#####################################################
#         TeamWon                   TeamLost
#####################################################
#list of Bans
#['champion','baron','dragon','riftHerald','tower'] kills
#gameDuration in mins
#gameVersion (used to determine the batch)
```

```python
#Step 4: extracting Match data for every match id
MatchIds = pd.read_csv("/content/MatchId.csv")
MatchIds_list = MatchIds["MatchId"]
#Data = pd.read_csv("Data.csv")
#matchData_list = df.values.tolist()
#matchData_list=[]

def getDataFromMatchByID(region, match_id,W):
  #Request Game Data from Api
  Game = W.match.by_id(region, match_id)
  InfoDto = Game['info']
  #Create Teams to collect champion and items
  TeamWon = []
  TeamLost = []
  #select the data collected for every player in the game
  PlayerStatsToAppend=['championId','item0', 'item1', 'item2', 'item3', 'item4', 'item5']
  #loop over all players
  for row in InfoDto['participants']:
    participants_row = [row[statName] for statName in PlayerStatsToAppend]
    winFlag=row['win']
    #if this player belongs to the winning team
    if winFlag:
      TeamWon.append(participants_row)
    #else this player belongs to the losing team
    else:
      TeamLost.append(participants_row)

  #Extract Meta Data for each team
  DataNames=['champion','baron','dragon','riftHerald','tower']
  for teamData in InfoDto['teams']:
    winFlag = teamData['win']
    objs = teamData['objectives']
    if winFlag:
      TeamWon.extend([[ban['championId'] for ban in teamData['bans']]])
      TeamWon.extend([[objs[DataName]['kills'] for DataName in DataNames]])
    else:
      TeamLost.extend([[ban['championId'] for ban in teamData['bans']]])
      TeamLost.extend([[objs[DataName]['kills'] for DataName in DataNames]])

  # create the follwing list Team won + its Data then team lost + its Data
  TeamWon.extend(TeamLost)
  # Add Game Duration in mins
  TeamWon.append([round(InfoDto['gameDuration']/(60),2)])
  # Add Game Version To Determin Batch at which the match occured
  TeamWon.append(InfoDto['gameVersion'][:5])
  #append a list of Strings
  matchData_list.append([','.join([str(elem) for elem in Data]) for Data in TeamWon])


columnNames=["Champ1Team1","Champ2Team1","Champ3Team1","Champ4Team1","Champ5Team1","BansTeam1","MetaTeam1",
      "Champ1Team2","Champ2Team2","Champ3Team2","Champ4Team2","Champ5Team2","BansTeam2","MetaTeam2",
      'gameDuration (Mins)','gameVersion']
for i in range(85001+2904,100015,len(keys)):
  time.sleep(1.5)
  print(len(matchData_list))
  for j in range(0,len(Ws)):
    getDataFromMatchByID(region, MatchIds_list[i+j], Ws[j])
    #check if data increased by 500 rows to save a back up
    if len(matchData_list)%500==0:
      df = pd.DataFrame(matchData_list, columns = columnNames)
      #save data in drive
      df.to_csv('Data1.csv',mode = 'w')
```
